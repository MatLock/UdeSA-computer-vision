```python
import requests
from io import BytesIO
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image

# 1. Load Pre-trained Model (e.g., BLIP-2)
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# 2. Load and Prepare Image from URL
img_url = 'https://www.militarykit.com/cdn/shop/files/mens-gildan-short-sleeve-heavy-cotton-tshirt-royal-blue.jpg?v=1766053652&width=1500'
response = requests.get(img_url)
image = Image.open(BytesIO(response.content)).convert('RGB')

# 3. Generate Description
inputs = processor(image, return_tensors="pt")
outputs = model.generate(**inputs)
description = processor.decode(outputs[0], skip_special_tokens=True)

print("Clothing Description:", description)
```
