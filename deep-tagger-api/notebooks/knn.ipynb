```python
import numpy as np
import matplotlib.pyplot as plt

from machine_learning.knn_model import KNNColor
```

```python
knn = KNNColor(colors_to_predict = 2)

def plot_colors(colors):
    plt.figure(figsize=(10, 2))
    for i, color in enumerate(colors):
        plt.fill_between([i, i+1], 0, 1, color=np.array(color)/255.0)
    plt.axis('off')
    plt.title('Dominant Colors Palette')
    plt.show()

```

```python
name, colors_list = knn.predict_k_colors('https://www.militarykit.com/cdn/shop/files/mens-gildan-short-sleeve-heavy-cotton-tshirt-royal-blue.jpg?v=1766053652&width=1500')
plot_colors(colors_list)
```
